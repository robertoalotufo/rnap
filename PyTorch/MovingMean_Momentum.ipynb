```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.arange(200)
y = np.random.normal(0.0,size=len(x)) + np.cos(2*np.pi/100 * x)
plt.plot(y);
```

```python
u_old = 0.
u = []
momentum = 0.9
for u_new in y:
    ua = (momentum * u_old) + (1-momentum) * u_new
    u_old = ua
    u.append(ua)

plt.plot(x,y,'r',x,u,'k');
```
