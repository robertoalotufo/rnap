# Média móvel - Momentum

A média móvel é utilizada em deep learning em duas situações principais: no cálculo da média do gradiente na otimização pelo gradiente descendente e no batch-norm, para estimar a média e a variância dos dados.

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## Geração sinal ruidoso

```python
i = np.arange(200)
u_signal = np.random.normal(0.0,size=len(i)) + np.cos(2*np.pi/100 * i)
plt.plot(u_signal);
```

## Filtrando com média móvel - $lambda$ - Momentum

$$ \boldsymbol{m}_{(t)} =  \lambda \boldsymbol{m}_{(t-1)} + (1 - \lambda) \boldsymbol{\mu} $$

O fato $\lambda$ é usualmente denominado de *momentum*. A seguir o sinal acima é filtrado com momentum de 90%.

Se significado é uma média móvel, onde o sinal de média filtrada é atualizada com 90% da média anterior e 10% com o novo sinal.

```python
m_old = 0.
m_signal = []
momentum = 0.9
for u in u_signal:
    m = (momentum * m_old) + (1-momentum) * u
    m_old = m
    m_signal.append(m)

plt.plot(i,u_signal,'r',i,m_signal,'k');
```
