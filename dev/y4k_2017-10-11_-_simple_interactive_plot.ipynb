```python
from IPython import display
import time
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.grid(True)
ax.set_ylim(0., 1.)

y = list()
line, = ax.plot(y)
while True:
    print(time.strftime('%F %T', time.localtime()))
    y.append(np.random.rand())
    line.remove()
    line, = ax.plot(y, 'r')
    time.sleep(1.)
    display.clear_output(wait=True)
    display.display(fig)
```
